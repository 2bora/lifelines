#*lifelines*: Survival Analysis in Python
##### Turtorial & Examples


*lifelines* gives a powerful tool to data analysts and statisticians looking for new methods to solve a common problem:
> How do I measure durations?

This question seems very vague and abstract, but thats only because we can so general in this space. Some more specific questions *lifelines* will help you solve are:

> How long can I expect a user to be on my service?
> Which treatment group has a longer expected lifetime?
> When is the next event likely to happen?
> Which factors influence durations of events?

The statistical tool that answers these question is Survival Analysis. *lifelines* is a implementation of survival analysis. What benefits does *lifelines* offer over other survival analysis implementations:

- built ontop of Pandas
- internal plotting library
- simple and intuitive classes and functions (*made for humans*)


```python
import numpy as np
np.set_printoptions(precision=2,  suppress=True) 
```

## Quick intro to Survival Analysis

Traditionally, survival analysis was developed to measure lifetimes of individuals. An actuary or health professional would ask a question like "how long does this population live for?", for example the population may be a nation's population (for actuaries), or a population sticken by a disease (the the medical professional's case). Sorta morbid. 

The anaylsis can be further applied to not just traditional *births and deaths*, but any duration. Medical professional might be interested in the *time between childbirths*, where a birth in this case is newborn's birth, and a death is becoming pregnant again! (obviously, we are loose with our definitions of *birth and death*) Another example is users on a service: a birth is a user who joins the service, and a death is when the user leaves the service. 

What makes measuring durations difficult is the passage of time itself! At the time you want to make inference on durations, it is possible, likely infact, that not all the death events have occured. These individuals in our population who have not been subject to the death event are labeled as *right-censored*: all the information we have on these individuals are thier current lifetime (which is naturally *less* than their acutal lifetimes). A common mistakes data analysts make is simply choosing to ignore the right-censorted individuals. We'll shall see why next.


Consider a case where the population is actually made up of two populations, $A$ and $B$. Population $A$ has a very low lifespan, say 2 time intervals on average, and population $B$ enjoys a much large lifespan, say 12 time intervals on average. Of course, we do not know this distinction before hand. Denote time $t$. At $t=10$, we wish to investigate the average lifespan. Below is an example of such a situation.

```python
figsize(12.5,6)

from lifelines.plotting import plot_lifetimes
from numpy.random import uniform, exponential

N = 25
current_time = 10
actual_lifetimes = np.array([[exponential(12), exponential(2)][uniform()<0.5] for i in range(N)])
observed_lifetimes = np.minimum(actual_lifetimes,current_time)
observed= actual_lifetimes < current_time

plt.xlim(0,25)
plt.vlines(10,0,30,lw=2, linestyles="--")
plt.xlabel('time')
plt.title('Births and deaths of our population, at $t=10$')
plot_lifetimes(observed_lifetimes, censorship=observed)
print "Observed lifetimes at time %d:\n"%(current_time), observed_lifetimes
```

The red lines denote the lifespan of instances where the death has been observed, and the blue lines denote the lifespan of the right-censored instances (deaths have not been observed). If we are asked to estimate the average lifetime of our population, and we naivly decided to *not* included the right-censored individuals, it is clear that we would be serverly underestimating the true average lifespan. 


Furthermore, if we instead simply took the mean of *all* observed lifespans, including the current lifespans of right-censored instances, we would *still* be underestimating the true average lifespan. Below we plot the actual lifetimes of all instances (though we do not see this information at $t=10$).

```python
plt.xlim(0,25)
plt.vlines(10,0,30,lw=2,linestyles="--")
plot_lifetimes(actual_lifetimes, censorship=observed)
```

Survival analysis was originally developed to solve this type of problem, that is, to deal with estimation when our data is right-censored, and we can't solely rely on historical data either. (The particular case when all death events have been made is much simpler, and statistics like sample averages are as the usual form once again -- though survival analysis is still very useful in this case as we shall later).

The observations need not always start at zero, either. This was done only for understanding in the above example. In the service example, where customers joining is a birth, a customer can enter observation at any time, and not necessarily at time 0.


We next introduce the two fundamental objects in survival analysis, the *survival function* and the *hazard function*.

-------

Let $T$ be the (possibly infinite, but always positive) random duration taken from the population under observation.  The survival function, $S(t)$, of a population is defined as 

$$ S(t) = Pr( T > t) $$

i.e., the probability the death event has not occured yet at time $t$. Note the following properties of the survival function:

2. $0 \le S(t) \le 1$
1. $S(0) = 1$
2. $F_T(t) = 1 - S(t)$, where $F_T(t)$ is the cumulative density function of $T$, which implies 
4.  $f_T(t) = -S'(t) $

We are also interested in the probability of dieing in the next instant, given we haven't expired yet. Mathematically, that is:

$$ \lim_{\delta t \rightarrow 0 } \; Pr( t \le T \le t + \delta t | T > t) $$

This quantity goes to 0 as $\delta t$ shrinks, so we divide this by the interval $\delta t$ (sorta like we do in calculus). This defines the hazard function at time $t$, $\lambda(t)$:

$$\lambda(t) =  \lim_{\delta t \rightarrow 0 } \; \frac{Pr( t \le T \le t + \delta t | T > t)}{\delta t} $$

It can be shown with quite elementary probability that this is equal to:

$$\lambda(t) = \frac{-S'(t)}{S(t)}$$

and solving this differential equation (yes, it is a differential equation):

$$S(t) = \exp\left( -\int_0^t \lambda(z) dz \right)$$


What I love about the above equation is that is defines **all** survival functions, and because the hazard function is arbitrary (i.e. it has no particular form), the entire function is non-parametric . Notice that we can now speak either about the survival function, $S(t)$, or the hazard function, $\lambda(t)$, and we can convert back and forth quite easily too. It also gives us another, albeit not very useful, expression for $T$: Upon differentiation and some algebra, we recover:

$$f_T(t) = \lambda(t)\exp\left( \int_0^t \lambda(z) dz \right)$$

Of course, we do not observe the true survival curve of a population. We must use the observed data to estimate it. We also want to continue to be non-parametric, that is not assume anything more about the data. The *best*\* method to recreate the survival function non-parametrically from the data is known as the Kaplan-Meier estimate. 

##### Example: Estimate $S(t)$ using the Kaplan-Meier estimate



```python
import pandas as pd
data = pd.read_csv('datasets/canadian_senators.csv', header = 0, index_col=0 )
```

```python
data[["Name", "start_date", "diff_days", "observed"]].head()
#the boolean columns `observed` refers to whether leaving office was observed or not. Senators 
# still in office have not been observed to have left.
```

From the `1ifelines.estimation` library, we'll need the `KaplanMeierFitter` for this exercise: 

```python
from lifelines.estimation import KaplanMeierFitter
kmf = KaplanMeierFitter()
```

For this estimation, we need the time each Senator was/has been in office, and whether or not they were observed to have left office (Senators still in office have not been observed to have left).

```python
T = data["diff_days"].values/365 #annualize days to years
C = data["observed"].values
```

Current, all arguments to *lifelines* estimators must be numpy arrays. We next use the `KaplanMeierFitter` method `fit` to fit the model to the data. (This is similar to, and was inspired by, another popular Python library [scikit-learn's](http://scikit-learn.org/stable/) fit/predict API)

```python
kmf.fit(T, censorship=C )
kmf.fit?
```

After calling the `fit` method, the `KaplanMeierFitter` has a field called `survival_function_`. (Again, we follow the styling of scikit-learn, and append an underscore to all estimated properties) The property is actually a pandas dataframe, so we can call `plot` on it:

```python
kmf.survival_function_.plot()
```

What do we see? The y-axis represents the probability a Senator is still around after $t$ years, where $t$ years is on the x-axis. We see that very few Senators make it past 40 years in office. The median time in office, which defines the point in time where on average 1/2 of the population has expired, is a property:

```python
kmf.median_
```

Interesting that it is only 13 years. Let's redefine our death event to being *only* retirement or resignation. For example, Willie Adams was a Senator between April 5th, 1977 and  June 22nd, 2009, and he eventually *retired* from office. These events define his particular birth and death: birth when joining, death when retiring.

Now consider Michael Adams, who was in office between January 7th, 1896 and January 1st, 1899. 
This latter date *was actually* his death: he died while still in office. Thus he never did retire, and we consider his senate retirement to be censored by his death.

(This is an example that has gladly redefined the birth and death events, and infact completely flips the idea upside down by using deaths as the censorship event. This is also an example where the current time is not the impeidment for complete observation -- there are alternative events (eg: death) that can stop an Senator for retiring. ) 

This type of analysis is called a competing-risk model, where there are multiple risks that can remove individuals from the population under study. The risk we are interested in is retirement/resignation, and the censorship event is death or unobserved. The reason the individual senator left office is in the `reason` column.

```python
C = ((data["reason"]=="Retirement") + (data["reason"] == "Resignation")).values
```

```python
kmf.fit(T, censorship=C, columns=["death_event=retirement/resignation"])
kmf.survival_function_.plot()
print kmf.median_
```

Thus, because we have factored in deaths and unobserved events, we can confidently say that the median time a Senator is in office is 25 years before retirement or resignition.

Actually, let's focus on just resignations. A senator rarely resigns (only 20% have), and it is often over a scandal. 

```python
C = (data["reason"] == "Resignation").values
kmf.fit(T, censorship=C, columns=["death_event=resignation"])
kmf.survival_function_.plot()
print kmf.median_
```

It looks like it starts quite flat, then drags down. If we were daring, we could make the presumption that Senators later in their term engage in more scandals (or make more mistakes).
