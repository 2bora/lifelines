#*lifelines*: Survival Analysis in Python
##### Documentation

*lifelines* gives a powerful tool to data analysts and statisticians looking for new methods to solve a common problem:
> How do I measure durations?

This question seems very vague and abstract, but thats only because we can so general in this space. Some more specific questions *lifelines* will help you solve are:

> How long can I expect a user to be on my service?
> Which treatment group has a longer expected lifetime?
> When is the next event likely to happen?
> Which factors influence durations of events?

The statistical tool that answers these question is Survival Analysis. *lifelines* is a implementation of survival analysis. What benefits does *lifelines* offer over other survival analysis implementations:

- built ontop of Pandas
- internal plotting library
- simple and intuitive classes and functions (*made for humans*)


## `lifelines.estimation`

In the `lifelines.estimation` sublibrary, there current exists three estimators. The API and styling was heavily inspired by the popular scikit-learn library. For example, estimators expose a `fit` method that accepts data, and same even expose a `predict` method. Also, quantities that are numerically estimated are appened with an underscore, eg: `survival_function_`. 

### `KaplanMeierFitter`

This class estimates the survival function, $S(t)$, of data with or without right-censored data.

    KaplanMeierFitter(alpha=0.95)

The variable `alpha` refers to the level of significance in the confidence interval. The class exposes the following methods:
   
    .fit(event_times, censorship=None, timeline=None, columns=['KM-estimate'] )
   
This method estimates the survival function of the survival data. The result is contained in the `survival_function_` property. 

- `event_times`: an (n,) numpy array of times (non-negative values) that the death event occured at. 
- `censorship`: an (n,) numpy array of booleans representing whether the individual in `event_times` was subject to the death event. If left as `None`, assumes all death events occured. 
- `timeline`: an numpy array of increasing times points of interest. This will form the basis of the estimated survival functions `index` (recall it is a DataFrame). For example, if `timeline=np.arange(10)`, then the returned DataFrame will only contain estimates at these points. If left as `None`, the index is the unique ascending elements of `event_times`. 
- `columns`: a length-1 list containing the column name of the returned survival function. 


**After** applying `fit` to the class, the object exposes the following properties and methods:

    .survival_function_

This is the Kaplan-Meier estimate of the survival function, contained in a DataFrame. (The index of the DataFrame is either the unique values in `event_times`, or `timeline`) 

    .median_

The median, possibly infinite, surival time of the population.

    .confidence_interval_

A (n,2) DataFrame, where n is the same size as `survival_function_'. It contains two columns, `upper_bound_<alpha>` and `lower_bound_<alpha>`, and the values represent the upper and lower confidence intervals respectively. 

    .plot(**kwargs)

 This methods plots the `survival_function_` and `confidence_interval_` on one figure. `kwargs` refers to any valid Matplotlib plotting arguments. 

### `NelsonAalenFitter`

The class estimates the *cumulative hazard function* over $t$, $\int_0^t h(z) dz $, from the data, with or without right-censored data.

    NelsonAalenFitter(alpha=0.95, nelson_aalen_smoothing=False)

The variable `alpha` refers to the level of significance in the confidence interval. `nelson_aalen_smoothing` is an alternative method to deal with ties (though the default method works equally as well).

The class exposes the following methods:
   
    .fit(event_times, censorship=None, timeline=None, columns=['NA-estimate'] )

This method estimates the Nelson-Aalen estimator of the cumulative hazard function, contained in the `cumulative_hazard_` property.

- `event_times`: an (n,) numpy array of times (non-negative values) that the death event occured at. 
- `censorship`: an (n,) numpy array of booleans representing whether the individual in `event_times` was subject to the death event. If left as `None`, assumes all death events occured. 
- `timeline`: an numpy array of increasing times points of interest. This will form the basis of the estimated cumulative hazards `index` (recall it is a DataFrame). For example, if `timeline=np.arange(10)`, then the returned DataFrame function will only contain estimates at these points. If left as `None`, the index is the unique ascending elements of `event_times`. 
- `columns`: a length-1 list containing the column name of the returned cumulative hazard function. 


**After** applying `fit` to the class, the object exposes the following properties and methods:

    .cumulative_hazard_

This is the Nelson-Aalen estimate of the cumulative hazard function, contained in a DataFrame. (The index of the DataFrame is either the unique values in `event_times`, or `timeline`) 

    .confidence_interval_

A (n,2) DataFrame, where n is the same size as `cumulative_hazard_'. It contains two columns, `upper_bound_<alpha>` and `lower_bound_<alpha>`, and the values represent the upper and lower confidence intervals respectively. 

    .plot(**kwargs)

 This methods plots the `cumulative_hazard_` and `confidence_interval_` on one figure. `kwargs` refers to any valid Matplotlib plotting arguments. 

### `AalenAdditiveFitter`

This class implements Aalen's additive covariate model. Specifically, we assume the hazard function looks like:

$$ \lambda(t) = \beta_0(t) + \beta_1(t)x_1 + ... + \beta_n(t)x_n $$

Traditionally, the estimates returned are the *cumulative coefficients*, i.e. we estimate the series

$$ \int_0^t \beta_i(z) dz, \;\;\; i=1..n $$

We estimate these quantities, and also perform kernel smoothing to recover $\beta_i(t)$.

Note: we currently have only implemented static covariates, and are working on adding the ability to have time-dependent covariates.

    AalenAddtiveFitter(fit_intercept=True, alpha=0.95, penalizer=0)

- `fit_intercept`: If False, ignores $\beta_0$. This is ill-advised unless the problem truely suggests it. 
- `alpha`: the alpha value for the confidence intervals.
- `penalizer`: The estimation of the coefficients at each time-step requires a linear regression step. The `penalizer` term represents adding an L2 penalizer to the regression (also known as ridge regression). This benefits the estimation in two ways: it makes the linear regression solvable, and controls the variability of the coefficients. Traditionally, this has not been included, but I'm adding it because it really helps.

The class exposes the following methods:

    fit(event_times, X, censorship=None, timeline=None, columns=None) 

Fits the data to Aalen additive model and creates a (n,d+1), or (n,d) in the case `fit_intercept=False`, DataFrame named `cumulative_hazards_`. 

- `event_times`: a (n,1) numpy array of the lifetimes of individuals under observation.
- `X`: a (n,d) numpy matrix or DataFrame that holds the (currently static) covariates. 
- `censorship`: an (n,) numpy array of booleans representing whether the individual in `event_times` was subject to the death event. If left as `None`, assumes all death events occured. 
- `timeline`:  an numpy array of increasing times points of interest. This will form the basis of the estimates `index` (recall it is a DataFrame). For example, if `timeline=np.arange(10)`, then the returned DataFrame will only contain estimates at these points. If left as `None`, the index is the unique ascending elements of `event_times`. 
- `columns`: a length-d list of names for the `cumulative_hazards` columns.
