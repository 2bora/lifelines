# Divorce Rates

```python
from lifelines.estimation import AalenAdditiveFitter
import pandas as pd
import numpy as np
%load_ext autoreload
%autoreload 2
```

```python
data = pd.read_csv('datasets/divorce.dat', sep="\s{2,10}")
```

```python
data.tail(10)
```

```python
del data["id"]
print data.shape
data = data.dropna()
```

```python
def to_bool(x):
    if x == "Yes":
        return 1
    elif x=="No":
        return 0
    else:
       return "error ", x
    
edu_mapping = { "None":0, "< 12 years": 1, "12-15 years": 2, "16+ years":3 }
def to_mapping(x, mapping = edu_mapping):
     try:
        return mapping[x]
     except:
         print "error: ", x
```

```python
data["heblack"] = data["heblack"].apply(to_bool, 1)
data["mixed"] = data["mixed"].apply(to_bool, 1)
data["div"] = data["div"].apply(to_bool, 1)
data["heduc"] = data["heduc"].apply(to_mapping)
```

```python
data["12-15 years"] = (data["heduc"] == 2).astype(float)
data["16+ years"] = (data["heduc"] == 3).astype(float)
#data["shewhite"] = ( data["heblack"]*data["mixed"] + (1-data["heblack"])*(1-data["mixed"]) == 1).astype(float)
data["bb"] = ( (data["heblack"])*(1-data["mixed"]) ).astype(float)
data["bw"] = ( (data["heblack"])*(data["mixed"]) ).astype(float)
data["ww"] = ( (1-data["heblack"])*(1-data["mixed"]) ).astype(float)
data["wb"] = ( (1-data["heblack"])*(data["mixed"]) ).astype(float)
```

```python
T = data["years"].values
C = data["div"].values
X = data[data.columns - ["heduc", "heblack", "years","div", "mixed"] ].values
```

```python
data.columns - ["heduc", "heblack", "years","div", "mixed"]
```

```python
aaf = AalenAdditiveFitter()
```

```python
aaf.fit(T[None,:], X, censorship=C, columns=["12-15 years","16+ years", "bb", "bw", "wb", "ww"])
```

```python
figsize(12.5, 8.5)
aaf.cumulative_hazards_.ix[:3200].plot()
```

```python
sv = (aaf.predict_survival_function(X[-50:,:]))
```
